### Importing important libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings 
warnings.filterwarnings('ignore')
```

### import dataset and extract dependent and independent variables

```python
df=pd.read_csv('Salary_Data.csv')
df
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,1].values
y
```

### visualize the dataset

```python
sns.distplot(df.YearsExperience,bins=10)
plt.show()
```

```python
sns.countplot(y=df.YearsExperience)
plt.show()
```

### splitting the dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

### Fitting the data to train

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

### Predicting the test set result

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
y_test
```

### visualize the training set

```python
plt.scatter(x_train,y_train,color='b')
plt.plot(x_train,lr.predict(x_train),color='r')
plt.title('Salary vs yearsofexperience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x_test,y_test,color='b')
plt.plot(x_test,lr.predict(x_test),color='r')
plt.title('Salary vs yearsofexperience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# cheak accurancy
from sklearn import metrics
print('MAE:',metrics.mean_absolute_error(y_test,y_pred))
print('MAE:',np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```
